1. Using the  titanic dataset, use the joint plot to visualize and analyze the chances of survival of the passengers in relation to their age

```python
#importing libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#import the data
data = sns.load_dataset("titanic")

data
```

```python
#analyzing the chances of survival in relation to their age
sns.jointplot(x = "age", y = "survived", kind = "scatter", alpha=0.1, data=data)#low alpha value to enhance transparency

plt.show()
```

**Interpretation of the graph

From the graph, we can see the relationship between age and survival. And we can infer that some of the older people in the age bracket between 50-70 died. The least survival being between 70-80. It also shows that some of the younger persons between 20-40 survived most.


Question2. Make a pie chart of a dataset containing at least 5 types of animals  in the feline family.(assign percentage numbers arbitrarily).

```python
#list six animal species of feline
feline = 'cheetah', 'puma', 'jaguar', 'leopard', 'lynx', 'tiger'
sizes = [10, 15, 15, 20, 30, 10]
explode = (0, 0, 0.1, 0, 0, 0) #explode the third slide


fig, ax = plt.subplots()
ax.pie(sizes, explode=explode, labels=feline, autopct='%1.0f%%',
       shadow=True, startangle=90)

ax.axis('equal')
plt.show()
```
